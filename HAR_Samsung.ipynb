```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ładowanie i przetwarzanie danych

```python
train_labels = pd.read_csv('samsung_train_labels.txt', header = None, sep = " ")
train_labels
```

```python
train_dataset = pd.read_csv('samsung_train.txt', header = None, sep = " ")
train_dataset
```

```python
train_dataset.drop(0, inplace=True, axis=1)
train_dataset
```

```python
features = pd.read_csv('features.txt', header = None, sep = " ")
features.drop(0, inplace=True, axis=1)
features.columns=['features']
features
```

```python
features_list = features['features'].tolist()
features_list
```

```python
train_dataset.columns = features_list
train_dataset
```

```python
test_dataset = pd.read_csv('samsung_test.txt', header = None, sep = " ")
test_dataset.drop(0, inplace=True, axis=1)
test_dataset.columns = features_list
test_dataset
```

```python
test_labels = pd.read_csv('samsung_test_labels.txt', header = None, sep = " ")
test_labels
```

```python
train_merged = pd.merge(train_labels,train_dataset,how='inner', right_index=True, left_index=True)
train_merged
```

```python
test_merged = pd.merge(test_labels,test_dataset,how='inner', right_index=True, left_index=True)
test_merged
```

```python
total_dataset = pd.concat([train_merged, test_merged])
total_dataset.reset_index(inplace=True)
total_dataset.drop('index', inplace=True, axis=1)
total_dataset
```

```python
total_dataset.rename(columns={0: 'activity'}, inplace=True)
total_dataset
```

```python
total_dataset['activity']=total_dataset['activity'].map({1: 'WALKING', 2: 'WALKING_UPSTAIRS', 3: 'WALKING_DOWNSTAIRS', 4: 'SITTING', 5: 'STANDING', 6: 'LAYING'})
total_dataset
```

```python
total_dataset.info(verbose=True, show_counts=True)
```

# Segmentacja zbioru za pomocą metody k-średnich

```python
X= total_dataset.drop('activity', axis=1)
y= pd.DataFrame(total_dataset['activity'])
y.columns=['y']
y
```

```python
from sklearn.cluster import KMeans

kmeans_per_k = [
    KMeans(n_clusters=k, max_iter=10, random_state=1).fit(X)
    for k in range(2, 10)]
```

```python
import matplotlib.pyplot as plt
inertias = [model.inertia_ for model in kmeans_per_k]
plt.plot(range(2, 10), inertias, 'bx-')
plt.xlabel('K')
plt.ylabel('SSE')
plt.title('Elbow Method using SSE')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = [
    silhouette_score(X, model.labels_)
    for model in kmeans_per_k
]
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.ylabel("Silhouette score", fontsize=14)
plt.grid()
plt.show()
```

```python
clf1 = KMeans(n_clusters=6, max_iter=100, random_state=1)

clf1.fit(X)
```

```python
y_pred1=pd.DataFrame(clf1.predict(X))
y_pred1.columns=['y_pred1']
y_pred1
```

```python
total_dataset = pd.concat([total_dataset, y_pred1], axis=1)
total_dataset
```

```python
summary = total_dataset.pivot_table(values='tBodyAcc-mean()-X', index='y_pred1', columns='activity', aggfunc='count', fill_value=0)

summary
```

```python
summary.plot(kind='bar', stacked=True)
plt.legend(loc=(1.05,0))
plt.ylabel('Observation count')
plt.xlabel('Clusters obtained in segmentation process')

plt.title('Actual activities distribution across segments')
```

Wnioski po segmentacji w oparciu o K-Means:
- Biorąc pod uwagę metodę łokcia, zarekomendowałbym 4 segmenty. OStatecznie zdecydowałem się na 6, żeby mieć tyle samo segmentów, co badanych aktywności.
- Metoda k-średnich bardzo słabo się sprawdziła, jeśli chodzi o segmentowanie aktywności. Różne aktywności są mocno wymieszane w ramach poszczególnych segmentów.
- Jedyną aktywnością, która w zdecydowanej większości została zgrupowana w jednorodny segment, to leżenie.
- Metoda k-średnich zupełnie nie była w stanie odróżnić siedzenia od stania. Podobny problem miał miejsce z chodzeniem, wchodzeniem po schodach oraz schodzeniem ze schodów.

# Segmentowanie za pomocą metody DBSCAN

```python
from sklearn.cluster import DBSCAN

clf2 = DBSCAN(eps=3.4, min_samples=5)

y_pred2 = pd.DataFrame(clf2.fit_predict(X))

y_pred2.columns=['y_pred2']

y_pred2
```

```python
total_dataset = pd.concat([total_dataset, y_pred2], axis=1)
total_dataset
```

```python
summary2 = total_dataset.pivot_table(values='tBodyAcc-mean()-X', index='y_pred2', columns='activity', aggfunc='count', fill_value=0)

summary2
```

```python
summary2.plot(kind='bar', stacked=True)
plt.legend(loc=(1.05,0))
plt.ylabel('Observation count')
plt.xlabel('Clusters obtained in segmentation process')

plt.title('Actual activities distribution across segments')
```

Wnioski po segmentacji w oparciu o DBSCAN:
- Metoda DBSCAN bardzo słabo się sprawdziła, jeśli chodzi o segmentowanie aktywności. 
- Wybrałem epsilon w taki sposób, żeby było jak najmniej obserwacji określonych jako outliery, ale i tak bardzo duża część obserwacji została w taki sposób określona.
- Metoda wyodrębniła jeden najbardziej liczny segment, w którym znajduje się większość obserwacji z aktywności leżenie, siedzenie i stanie.
- Porównując tę metodę to k-średnich, to metoda k-średnich była przynajmniej w stanie wyodrębnić do jednego segmentu leżenie.

Wnioski po segmentacji w oparciu o GMM:

- Wśród 6 segmentów, mamy 3 pary podobnych segmentów: są dwa segmenty, gdzie siedzenie i stanie są mylone, są dwa segmenty zawierające mieszankę chodzenia, wchodzenia po schodach i schodzenia po schodach, oraz dwa segmenty z leżeniem.
- O tyle GMM jest w tym wypadku lepsze niż k-średnich, że w ramach tych 3 par aktywności są jednorodne.


# Klasyfikacja supervised learning

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold

X=np.array(X)
y=np.array(y)

kf = KFold(n_splits=5, shuffle=True, random_state=1)

models = []
scores = []

for fold_nr, (train_idx, test_idx) in enumerate(kf.split(X)):
  X_train = X[train_idx]
  X_test =X[test_idx]

  y_train = y[train_idx]
  y_test = y[test_idx]

  clf = RandomForestClassifier(n_estimators=100, max_depth=3, min_samples_leaf=5).fit(X_train, y_train)

  models.append(clf)
  scores.append(clf.score(X_test, y_test))


print("wyniki poszczególnych foldów: ", scores)
print("średni wynik wszystkich foldów: ", np.array(scores).mean())

```

# Dodatkowe odpowiedzi / uwagi na pytania z zadania:

- Standaryzacja jest niepotrzebna - dane na wejściu są wystandaryzowane.
- W przypadku metod segmentacyjnych należących do unsupervised learning, nie ma potrzeby podziału danych na treningowe i testowe. Taka potrzeba oczywiście była w przypadku klasyfikacji, i tam podział na te zbiory miał miejsce przy okazji kroswalidacji.
- Czym różni się KMeans, K-fold i K Nearest Neighbors? Metoda k-średnich to metoda segmentacji należąca do unsupervised learning. Metoda k najbliższych sąsiadów to metoda prognozowania, mająca zastosowanie w klasyfikacji jak również regresji nieparametrycznej. K-fold to metoda optymalizacji modeli dzięki wykorzystaniu kroswalidacji, która zapobiega zjawisku przeuczeniu się modeli.

- Oczywiście warto byłoby wykorystać PCA przed wykonaniem segmentacji.
